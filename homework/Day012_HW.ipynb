### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup as bf

url = 'https://www.ettoday.net/news/news-list.htm'

result = requests.get(url)

#print(result.text)

soup = bf(result.text)

#print(soup.prettify())
#print('------------------')
result = soup.find_all('h3')
for item in result:
    temp = str(item.get_text(separator=" ")).split(' ')
    if len(temp)  >=  4:
         print(item.get_text(separator=" "))



```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python

all_new = list()
for item in result:
    temp = str(item.get_text(separator=" ")).split(' ')
    if len(temp)  >=  4:
        single_new = dict()
        single_new[temp[2]] = temp[-1]
        print(single_new)
        all_new.append(single_new)
         #print(item.get_text(separator=" "))


#print(all_new)
```

### ③ 哪一個類別的文章最多

```python
title_list = list()
for item in all_new:
    for key ,value in item.items():
        title_list.append(key)
        
max_title = max(title_list,key=title_list.count)   

print(max_title)
```

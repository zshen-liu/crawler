### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
import xmltodict
with open('./example/64_72hr_CH.xml', 'r') as fn:
    xml = fn.read()
    result = dict(xmltodict.parse(xml))
    location = list()
    for item in result['cwbopendata']['dataset']['locations']['location'] :
        if item['locationName']:
            location.append(item['locationName'])
        else:
            pass
    print(location)
    print(len(location))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

import xmltodict
with open('./example/64_72hr_CH.xml', 'r') as fn:
    xml = fn.read()
    result = dict(xmltodict.parse(xml))
    location_temp = dict()
    for weather_elements in result['cwbopendata']['dataset']['locations']['location']:
        print(weather_elements['locationName'])
        for items in weather_elements['weatherElement']:
            if items['elementName'] == 'T':
                print("時間: {0}".format(items['time'][0]['dataTime']))
                print("溫度: {0}".format(items['time'][0]['elementValue']['value']))
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

import xmltodict
with open('./example/64_72hr_CH.xml', 'r') as fn:
    xml = fn.read()
    result = dict(xmltodict.parse(xml))
    location_temp = dict()
    for weather_elements in result['cwbopendata']['dataset']['locations']['location']:
        print(weather_elements['locationName'])
        for items in weather_elements['weatherElement']:
            if items['elementName'] == 'T':
                for item in items['time']:
                    print("時間: {0}".format(item['dataTime']))
                    print("溫度: {0}".format(item['elementValue']['value']))
```

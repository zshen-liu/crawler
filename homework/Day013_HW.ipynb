### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup as bf

result = requests.get('https://www.ptt.cc/bbs/NBA/index.html')


soup = bf(result.text)

result = soup.find_all('div',class_='r-ent')
for item in result:
    #print(item)
    title = item.find(class_='title').text.strip()
    author= item.find(class_='author').text.strip()
    date = item.find(class_='date').text.strip()
    if author != '-':
        print('title is {0} , author is {1} , date is {2}'.format(title,author,date))
```

### ③ 試著爬爬看其他版的文章

```python
import requests
from bs4 import BeautifulSoup as bf

host = 'https://www.ptt.cc'
last_page = '/bbs/NBA/index.html'

url = host+last_page
count = 3
for i in range(1,count):
    print("the i is {0}".format(i))
    result = requests.get(url)
    soup = bf(result.text)
    page_list = soup.find_all('a',class_='btn wide')
    #print(page_list)
    up_page = page_list[1].get('href')
    result = soup.find_all('div',class_='r-ent')
    for item in result:
        title = item.find(class_='title').text.strip()
        author= item.find(class_='author').text.strip()
        date = item.find(class_='date').text.strip()
        if '刪除' not in title:
            print('title is {0} , author is {1} , date is {2}'.format(title,author,date))
            
        url = host+up_page
```
